# Example on using the fluxSort() function to understand FBA output

```python
import cobra, math, pandas as pd, numpy as np
from cobra.solvers import get_solver_name
from cobra import Model, Reaction, Metabolite
from cobra.flux_analysis import parsimonious
from fluxSort import *

# Some options for pandas to show output nicely
pd.set_option('display.max_rows', 1000) # Show everything
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 200)
```

```python
# Load Recon2
model = cobra.io.load_matlab_model('Recon2.v03_pythonCompatible.mat')
```

### Below an example of what you would want to do in practice with fluxSort

```python
# Choose you objective function, define which exchange reactions should be on/off etc.
# below for the plain Recon2 case on biomass production

# Perform parsimonious FBA (or regular FBA but that returns more messy results ;) )
# Note that you have to load the right modules up above
pfbaSol = cobra.flux_analysis.parsimonious.optimize_minimal_flux(model)

# this is where you call fluxSort by giving it the model and FBA solution.
# the function returns a table (actually a 'dataframe') object which you can print
# On the right side of the table in the 'Cat' column you see the function's attempt to describe what this reaction is doing. 
# THIS IS NOT PERFECT; but seems helpful so far. 
# If you see weird stuff please let me know so I can improve it :)
fbaTableSorted = fluxSort(model,pfbaSol)
fbaTableSorted
```
